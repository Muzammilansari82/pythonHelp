Pivot


```python
df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two','two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
df
```

```python
df.pivot(index='foo', columns='bar', values='baz')
```

```python
df = pd.DataFrame({
    'a' : ['one','one','one','two','two','two'],
    'b' : ['A', 'B', 'C', 'A', 'B', 'C'],
    'c' : [1, 2, 3, 4, 5, 6],
    'd' :['x', 'y', 'z', 'q', 'w', 't']
})
df
```

```python
df.pivot(index = 'a', columns= 'b',values = 'c')
```

```python
df = pd.DataFrame({
       "lev1": [1, 1, 1, 2, 2, 2],
       "lev2": [1, 1, 2, 1, 1, 2],
       "lev3": [1, 2, 1, 2, 1, 2],
       "lev4": [1, 2, 3, 4, 5, 6],
       "values": [0, 1, 2, 3, 4, 5]})
df
```

```python
df.pivot(index = 'lev4' , columns = 'lev3', values = 'values')
```

DRY RUN

```python
df.pivot(index = 'lev1',columns = ['lev2','lev3','lev4'],values='values')
```

Pivot Table

```python
df = pd.DataFrame({"A": ["foo", "foo", "foo", "foo", "foo",
                         "bar", "bar", "bar", "bar"],
                   "B": ["one", "one", "one", "two", "two",
                         "one", "one", "two", "two"],
                   "C": ["small", "large", "large", "small",
                         "small", "large", "small", "small",
                         "large"],
                   "D": [1, 2, 2, 3, 3, 4, 5, 6, 7],
                   "E": [2, 4, 5, 5, 6, 6, 8, 9, 9]})
df
```

DRY RUN

columns     small  large

a    b
foo   one  1 2

      two  3 n


bar   one  5  4

      two  6  7

```python
import numpy as np
```

```python
table = pd.pivot_table(df, values='D', index=['A', 'B'],columns=['C'])
table
```

```python
table = pd.pivot_table(df, values='D', index=['A', 'B'],
                       columns=['C'], aggfunc=np.sum)
table
```

```python
table = pd.pivot_table(df, values='D', index=['A', 'B'],
                       columns=['C'], aggfunc=np.mean)
table
```

```python
pd.pivot_table(df,values = 'D' , index = ['A','B'],columns =['C'],aggfunc=np.sum, fill_value = 0)
```
